## 5. Data Cleaning - Remove Missing Columns

```python
# Remove columns with 100% missing values
cols_to_drop = missing_df[missing_df['결측치 비율(%)'] == 100.0].index.tolist()
print(f"Columns with 100% missing values: {len(cols_to_drop)}")
print(cols_to_drop)

# Also remove kepler_name
if 'kepler_name' not in cols_to_drop:
    cols_to_drop.append('kepler_name')

print(f"\nTotal columns to drop: {len(cols_to_drop)}")

# Create cleaned dataset
kepler_clean = kepler.drop(columns=cols_to_drop)

print(f"\nOriginal shape: {kepler.shape}")
print(f"Cleaned shape: {kepler_clean.shape}")
print(f"Removed {len(cols_to_drop)} columns")

# Check remaining missing values
remaining_missing = kepler_clean.isnull().sum()
remaining_missing = remaining_missing[remaining_missing > 0].sort_values(ascending=False)
print(f"\nColumns with remaining missing values: {len(remaining_missing)}")
print(f"\nTop 10 columns with missing values:")
print(remaining_missing.head(10))
```

## 6. Correlation Analysis

```python
# Select only numeric columns for correlation analysis
numeric_cols_clean = kepler_clean.select_dtypes(include=[np.number]).columns
print(f"Number of numeric columns: {len(numeric_cols_clean)}")

# Calculate correlation matrix
corr_matrix = kepler_clean[numeric_cols_clean].corr()

# Plot correlation heatmap for all numeric features
plt.figure(figsize=(20, 16))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8},
            fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix - All Numeric Features', fontsize=16, weight='bold', pad=20)
plt.tight_layout()
plt.show()

print(f"\nCorrelation matrix shape: {corr_matrix.shape}")
```

```python
# Find highly correlated feature pairs (excluding diagonal)
# Create a copy of correlation matrix with upper triangle set to NaN
corr_upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find correlations above 0.8 (high correlation)
high_corr = []
for column in corr_upper.columns:
    high_corr_features = corr_upper[column][abs(corr_upper[column]) > 0.8].sort_values(ascending=False)
    for idx, value in high_corr_features.items():
        high_corr.append({'Feature 1': column, 'Feature 2': idx, 'Correlation': round(value, 3)})

high_corr_df = pd.DataFrame(high_corr)
print("=" * 80)
print(f"Highly Correlated Feature Pairs (|correlation| > 0.8): {len(high_corr_df)}")
print("=" * 80)
if len(high_corr_df) > 0:
    print(high_corr_df.head(30))
else:
    print("No highly correlated pairs found.")
```

## 7. Key Features Distribution by Class

```python
# Select key features for visualization
key_features = ['koi_period', 'koi_prad', 'koi_teq', 'koi_insol', 'koi_steff', 'koi_srad']

# Filter features that exist in the cleaned dataset and don't have all missing values
available_features = [f for f in key_features if f in kepler_clean.columns and kepler_clean[f].notna().any()]

print(f"Available key features: {available_features}")
print(f"Total: {len(available_features)}")

# Create box plots for each feature grouped by disposition
n_features = len(available_features)
n_cols = 3
n_rows = (n_features + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 5))
axes = axes.flatten() if n_features > 1 else [axes]

for idx, feature in enumerate(available_features):
    ax = axes[idx]
    
    # Create box plot
    kepler_clean.boxplot(column=feature, by='koi_disposition', ax=ax, 
                         patch_artist=True, showfliers=False)
    
    ax.set_title(f'{feature} by Disposition', fontsize=12, weight='bold')
    ax.set_xlabel('Disposition', fontsize=10)
    ax.set_ylabel(feature, fontsize=10)
    ax.get_figure().suptitle('')  # Remove default title
    
# Hide extra subplots
for idx in range(n_features, len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Distribution plots for key features by class
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten() if n_features > 1 else [axes]

colors = {'CANDIDATE': '#e74c3c', 'CONFIRMED': '#3498db', 'FALSE POSITIVE': '#2ecc71'}

for idx, feature in enumerate(available_features):
    ax = axes[idx]
    
    for disposition in kepler_clean['koi_disposition'].unique():
        data = kepler_clean[kepler_clean['koi_disposition'] == disposition][feature].dropna()
        ax.hist(data, alpha=0.5, label=disposition, bins=30, color=colors.get(disposition, 'gray'))
    
    ax.set_xlabel(feature, fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)
    ax.set_title(f'{feature} Distribution by Disposition', fontsize=12, weight='bold')
    ax.legend()
    ax.grid(alpha=0.3)

# Hide extra subplots
for idx in range(n_features, len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
```

## 9. Key Findings Summary

### Dataset Overview
- **Original Dataset**: 9,564 rows × 141 columns
- **Cleaned Dataset**: 9,564 rows × 121 columns (removed 20 columns)
- **Removed Columns**: 19 columns with 100% missing values + `kepler_name`
- **Numeric Features**: 105 columns

### Class Distribution (Imbalanced)
- **FALSE POSITIVE**: 4,839 (50.6%) ⚠️
- **CONFIRMED**: 2,746 (28.7%)
- **CANDIDATE**: 1,979 (20.7%)

### Correlation Insights
- Found **89 highly correlated feature pairs** (|r| > 0.8)
- Many error columns are perfectly correlated with their base features
- Strong correlations found:
  - `koi_period` ↔ `koi_dor` (r = 0.989)
  - `koi_period` ↔ `koi_sma` (r = 0.872)
  - `koi_ror` ↔ `koi_impact` (r = 0.994)

### Feature Characteristics by Class
Key observations from box plots and distributions:
- **koi_period**: CANDIDATE shows wider distribution
- **koi_prad**: Similar distributions across classes, mostly small values
- **koi_teq**: FALSE POSITIVE tends to have higher equilibrium temperature
- **koi_insol**: Highly skewed, concentrated at low values
- **koi_steff**: Stellar effective temperature relatively consistent
- **koi_srad**: Stellar radius shows similar patterns

### Recommendations for Model Building
1. **Handle class imbalance**: Use SMOTE, class weights, or stratified sampling
2. **Feature selection**: Remove highly correlated features to avoid multicollinearity
3. **Handle missing values**: Still have 100 columns with missing data
4. **Feature scaling**: Required due to different scales (e.g., period vs radius)
5. **Feature engineering**: Consider creating interaction terms or polynomial features
